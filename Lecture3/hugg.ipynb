# Hugging Face Summarization Models

This notebook demonstrates how to use pre-trained models from Hugging Face for text summarization.

```python
# Install required packages (if not already installed)
# !pip install transformers torch
```

```python
from transformers import pipeline

# Initialize the summarization pipeline
# Using facebook/bart-large-cnn - one of the best summarization models
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

print("Model loaded successfully!")
```

```python
# Example text to summarize
article = """
The field of natural language processing has seen tremendous growth in recent years, 
driven by advances in deep learning and the availability of large datasets. 
Transformer models, introduced in 2017, have revolutionized the field by enabling 
better performance on a wide range of NLP tasks including machine translation, 
text generation, and question answering. Models like BERT, GPT, and T5 have set 
new benchmarks across multiple tasks. These models use self-attention mechanisms 
to process input sequences in parallel, making them more efficient than previous 
recurrent neural network architectures. The pre-training and fine-tuning paradigm 
has become the standard approach, where models are first trained on large amounts 
of unlabeled text and then fine-tuned on specific downstream tasks with smaller 
labeled datasets. This transfer learning approach has democratized NLP, allowing 
researchers and practitioners to achieve state-of-the-art results without requiring 
massive computational resources for training from scratch.
"""

# Generate summary
summary = summarizer(article, max_length=130, min_length=30, do_sample=False)

print("Original text length:", len(article.split()))
print("\nSummary:")
print(summary[0]['summary_text'])
print("\nSummary length:", len(summary[0]['summary_text'].split()))
```

## Alternative Summarization Models

Popular models for summarization:
- **facebook/bart-large-cnn**: BART model fine-tuned on CNN/DailyMail (best quality)
- **sshleifer/distilbart-cnn-12-6**: Distilled BART (faster, smaller)
- **google/pegasus-cnn_dailymail**: PEGASUS model
- **t5-small** or **t5-base**: T5 models (versatile)
- **philschmid/bart-large-cnn-samsum**: Fine-tuned for conversation summarization

```python
# Example with a smaller, faster model
print("Loading smaller model for faster inference...")
summarizer_fast = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# Shorter text example
short_text = """
Artificial intelligence is transforming healthcare by enabling early disease detection, 
personalized treatment plans, and improved patient outcomes. Machine learning algorithms 
can analyze medical images, predict patient risks, and assist doctors in making more 
accurate diagnoses. However, challenges remain in ensuring data privacy, model 
interpretability, and equitable access to these technologies.
"""

summary_fast = summarizer_fast(short_text, max_length=50, min_length=20, do_sample=False)

print("\nOriginal:", short_text.strip())
print("\nSummary:", summary_fast[0]['summary_text'])
```

```python
# Customizing summarization parameters
def custom_summarize(text, max_len=100, min_len=30, length_penalty=2.0, num_beams=4):
    """
    Custom summarization with adjustable parameters
    
    Parameters:
    - max_length: Maximum length of summary
    - min_length: Minimum length of summary
    - length_penalty: Penalty for length (higher = shorter summaries)
    - num_beams: Beam search size (higher = better quality but slower)
    """
    summary = summarizer(
        text,
        max_length=max_len,
        min_length=min_len,
        length_penalty=length_penalty,
        num_beams=num_beams,
        early_stopping=True
    )
    return summary[0]['summary_text']

# Test custom function
result = custom_summarize(article, max_len=80, min_len=40, num_beams=4)
print("Custom summary:")
print(result)
```

```python
import numpy as np

# ----------------------------
# Step 1: Define "embeddings"
# ----------------------------
# 3 tokens, embedding dimension = 4
X = np.array([
    [1.0, 0.0, 1.0, 0.0],  # token 1 (e.g., "He")
    [0.0, 1.0, 0.0, 1.0],  # token 2 (e.g., "ate")
    [1.0, 1.0, 0.0, 0.0]   # token 3 (e.g., "apple")
])

# ----------------------------
# Step 2: Define projection matrices
# ----------------------------
np.random.seed(42)

W_Q = np.random.rand(4, 4)
W_K = np.random.rand(4, 4)
W_V = np.random.rand(4, 4)

# ----------------------------
# Step 3: Compute Q, K, V
# ----------------------------
Q = X @ W_Q
K = X @ W_K
V = X @ W_V

print("Q:\n", Q)
print("\nK:\n", K)
print("\nV:\n", V)

# ----------------------------
# Step 4: Compute attention scores
# ----------------------------
scores = Q @ K.T #Transpose K to get dot product between Q and K (all keys (all tokens in the sequence))

# Scale 
d_k = K.shape[1]
scores = scores / np.sqrt(d_k)

print("\nAttention Scores:\n", scores)

# ----------------------------
# Step 5: Softmax
# ----------------------------
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

weights = softmax(scores)

print("\nAttention Weights:\n", weights)

# ----------------------------
# Step 6: Compute output
# ----------------------------
# weighted sum of V using attention weights
output = weights @ V

print("\nFinal Output (Contextualized Representations):\n", output)
```

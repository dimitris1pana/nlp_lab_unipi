```python
import numpy as np
import matplotlib.pyplot as plt
# demonstrating how biases can shift neural outputs before activation
# Define vectors
a = np.array([3, 3])          # Orange vector from origin
b = np.array([-4, -4])        # Orange vector added to a
sum_vector = a + b            # Resultant vector ([-1, 2])

# Plot
fig, ax = plt.subplots()
ax.set_title("Vector/Matrix Addition")
ax.set_xlim(-5, 5)
ax.set_ylim(-2, 4)
ax.grid(True)

# Plot base point
ax.plot(0, 0, 'ko')  # origin dot

# Plot vector a (from origin)
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='orange')
ax.text(a[0] + 0.1, a[1], f'{a}', fontsize=10)

# Plot vector b (from origin)
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='darkorange')
ax.text(b[0] - 1.5, b[1] - 0.2, f'{b}', fontsize=10)

# Plot vector b starting from the tip of a
ax.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='orangered')
ax.text(sum_vector[0] - 1.2, sum_vector[1], f'{sum_vector}', fontsize=10, color='orangered')

# Vector b from tip of a (tip-to-tail)
ax.quiver(*a, *b, angles='xy', scale_units='xy', scale=1, color='orangered')
ax.text(sum_vector[0] - 1.2, sum_vector[1], f'{sum_vector}', fontsize=10, color='orangered')

# Draw line from origin to sum_vector (resultant vector)
ax.plot([0, sum_vector[0]], [0, sum_vector[1]], color='black', linestyle='--', linewidth=1.5)
ax.text(sum_vector[0] + 0.1, sum_vector[1] - 0.4, 'Resultant Vector', fontsize=10, color='black')

# Plot result (a + b)
ax.plot(sum_vector[0], sum_vector[1], 'ro')  # red dot at end
ax.axvline(x=sum_vector[0], linestyle='--', color='gray')
ax.axhline(y=sum_vector[1], linestyle='--', color='gray')

# Display the expression
plt.text(-4.5, -1.5, f"{a} + {b} = {sum_vector}", fontsize=11, color='black')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
